```python
import requests
import pandas as pd
```

```python
# -----------------------------
# Configuración API
# -----------------------------

BASE_URL = "https://api.data.gov/ed/collegescorecard/v1/schools"

API_KEY = "DEMO_KEY"  # Luego pueden cambiarla por una real

params = {
    "api_key": API_KEY,
    "school.name": "Austin Community College",
    "fields": ",".join([
        "school.name",
        "school.city",
        "school.state",
        "latest.student.size",
        "latest.finance.total_revenue",
        "latest.cost.avg_net_price.public",
        "school.ownership"
    ]),
    "per_page": 100
}

# Llamada a la API
response = requests.get(BASE_URL, params=params)

print("Status:", response.status_code)

data = response.json()
results = data["results"]

df = pd.DataFrame(results)

df.head()
```

```python
# -----------------------------
# Limpieza básica
# -----------------------------

# Eliminar filas vacías
df = df.dropna(how="all")

# Rellenar valores nulos
df = df.fillna(0)

# Quitar duplicados
df = df.drop_duplicates()

print("Datos limpios:", df.shape)
```

```python
# -----------------------------
# Guardar CSV
# -----------------------------

df.to_csv("education_data_ACC.csv", index=False)

print("Archivo guardado correctamente")
```
